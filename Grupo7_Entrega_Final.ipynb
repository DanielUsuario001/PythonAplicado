## Sección 1: NumPy - Manipulación Avanzada de Arrays

**1.1.** Crea un array de NumPy llamado `vector_base` que contenga una secuencia de números desde 10 hasta 115 (inclusive), con un paso de 3 entre cada número.

```python
import numpy as np #libreria

vector_base = np.arange(10, 116, 3)

print("Vector base (Resultado):", vector_base.tolist())
print("Longitud (Resultado):", len(vector_base))
```

**1.2.** A partir del `vector_base` anterior, crea una matriz de 5x7 llamada `matriz_A`.

```python
matriz_A = vector_base[:35].reshape(5, 7)

print("Matriz A 5x7 (Resultado):\n", matriz_A)
```

**2.1.** Crea una segunda matriz de 5x7 llamada `matriz_B` que contenga números enteros aleatorios entre 1 y 10.

**2.2.** Realiza una multiplicación elemento por elemento (element-wise) entre `matriz_A` y `matriz_B` y almacena el resultado en `matriz_C`.

```python
rng = np.random.default_rng(42)
matriz_B = rng.integers(1, 11, size=(5, 7))
matriz_C = matriz_A * matriz_B

print("Matriz B (Resultado):\n", matriz_B)
print("Matriz C = A * B (Resultado):\n", matriz_C)
```

**3.1.** Utilizando slicing, selecciona una submatriz de 2x2 de la esquina inferior derecha de `matriz_C`.

**3.2.** Calcula el promedio de esta submatriz.

```python
submatriz = matriz_C[3:, 5:]  # filas 4–5 y columnas 6–7
promedio = submatriz.mean()

print("Submatriz (Resultado):\n", submatriz)
print("Promedio de la submatriz (Resultado):", round(float(promedio), 2))
```

**5.** Hemos notado que hay un valor faltante (`NaN`) en la columna `Tasa_Alfabetizacion`. Imputa este valor faltante utilizando la **mediana** de esa misma columna.

```python
mediana_alf = df_desarrollo['Tasa_Alfabetizacion'].median()
df_desarrollo['Tasa_Alfabetizacion'] = df_desarrollo['Tasa_Alfabetizacion'].fillna(mediana_alf)

print("Mediana usada (Resultado):", mediana_alf)
df_desarrollo[['Pais','Tasa_Alfabetizacion']]
```

**7.** Crea una nueva columna condicional `Categoria_Desarrollo`. Clasifica a los países según las siguientes reglas:
- 'Alto': Si `Esperanza_de_vida` > 80 Y `PBI_per_capita` > 25000.
- 'Medio-Alto': Si `Esperanza_de_vida` > 76 Y `PBI_per_capita` > 8000.
- 'Medio': Para todos los demás casos.

```python
condiciones = [
    (df_desarrollo['Esperanza_de_vida'] > 80) & (df_desarrollo['PBI_per_capita'] > 25000),
    (df_desarrollo['Esperanza_de_vida'] > 76) & (df_desarrollo['PBI_per_capita'] > 8000)
]

elecciones = ['Alto', 'Medio-Alto']
df_desarrollo['Categoria_Desarrollo'] = np.select(condiciones, elecciones, default='Medio')
df_desarrollo[['Pais','Categoria_Desarrollo']]
```

**8.1.** Filtra el DataFrame para encontrar los países de 'Sudamérica' con un `Indice_Gini` menor a 45.

**8.2.** Ordena el resultado del filtro anterior por `PBI_per_capita` de forma descendente.

```python
sud_gini = df_desarrollo[(df_desarrollo['Continente'] == 'Sudamérica') & (df_desarrollo['Indice_Gini'] < 45)]
sud_gini_ord = sud_gini.sort_values('PBI_per_capita', ascending=False)
sud_gini_ord.reset_index(drop=True)
```

**9.** Utilizando `groupby()`, calcula para cada `Continente`:
- La **media** de `Esperanza_de_vida`.
- El **máximo** de `Indice_Gini`.
- El **conteo** de países.

```python
resumen_continente = (
    df_desarrollo.groupby('Continente', as_index=True)
      .agg(media_Esperanza_de_vida=('Esperanza_de_vida','mean'),
           max_Indice_Gini=('Indice_Gini','max'),
           conteo_paises=('Pais','count'))
      .round({'media_Esperanza_de_vida':2})
)
resumen_continente
```

**14.1.** Calcula la matriz de correlación para las columnas numéricas del DataFrame.

**14.2.** **Interpretación:** Identifica el par de variables con la correlación positiva más fuerte y el par con la correlación negativa más fuerte (excluyendo la correlación de una variable consigo misma). ¿Tienen sentido estos resultados en el mundo real?

```python
corr = df_desarrollo.select_dtypes(include='number').corr().round(3)
print("Matriz de correlación (Resultado):")
display(corr)


corr_vals = corr.abs().where(~np.eye(len(corr), dtype=bool))
max_val = corr_vals.unstack().dropna().sort_values(ascending=False).head(1)
mejor_par, mejor_r = max_val.index[0], float(max_val.iloc[0])
print("Par más correlacionado:", mejor_par, "con r =", mejor_r)
```

*Escribe tu interpretación aquí...*

La correlación más alta es entre el PBI per cápita y la esperanza de vida, lo que muestra que los países con más ingreso por persona suelen tener más años de vida. La más negativa es entre el PBI per cápita y el índice de Gini, lo que indica que a mayor ingreso normalmente hay menos desigualdad. Esto coincide con lo que pasa en la vida real.
